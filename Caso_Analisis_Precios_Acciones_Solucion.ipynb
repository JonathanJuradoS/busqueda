# üìà Caso Pr√°ctico: An√°lisis de Precios de Acciones

## üß© Contexto:
En el √°mbito financiero, las empresas y los inversores monitorean constantemente la evoluci√≥n del precio de sus acciones para tomar decisiones estrat√©gicas de compra y venta. El an√°lisis t√©cnico se basa en patrones de comportamiento del mercado, y uno de los puntos clave es la identificaci√≥n de **m√°ximos y m√≠nimos locales** en series temporales.

Este caso simula el an√°lisis de los precios diarios de una acci√≥n durante 60 d√≠as consecutivos. A trav√©s de los algoritmos de detecci√≥n de extremos locales, se busca identificar:
- **Zonas de compra (m√≠nimos locales)**: donde el precio ha descendido respecto a sus d√≠as vecinos.
- **Zonas de venta (m√°ximos locales)**: donde el precio ha alcanzado un pico en comparaci√≥n con sus d√≠as cercanos.

El an√°lisis se complementar√° con visualizaciones para facilitar la interpretaci√≥n de los resultados.

---

## üéØ Objetivo:
1. Cargar y explorar datos reales simulados desde un archivo CSV.
2. Visualizar la evoluci√≥n temporal de los precios.
3. Aplicar l√≥gica de b√∫squeda local para encontrar m√°ximos y m√≠nimos.
4. Marcar gr√°ficamente los puntos estrat√©gicos de compra y venta.
5. Reflexionar sobre el uso de IA en decisiones de inversi√≥n.


```python

import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv("/content/Dataset_Precios_Acciones.csv", parse_dates=["Fecha"])
df.head()
```

```python

# Visualizar evoluci√≥n de precios
plt.figure(figsize=(12,6))
plt.plot(df["Fecha"], df["Precio"], label='Precio de la Acci√≥n', color='blue')
plt.title("Evoluci√≥n del Precio de la Acci√≥n")
plt.xlabel("Fecha")
plt.ylabel("Precio")
plt.grid(True)
plt.legend()
plt.show()
```

```python

# Detectar m√°ximos y m√≠nimos locales
def detectar_extremos(precios):
    maximos = []
    minimos = []
    for i in range(1, len(precios)-1):
        if precios[i] > precios[i-1] and precios[i] > precios[i+1]:
            maximos.append(i)
        elif precios[i] < precios[i-1] and precios[i] < precios[i+1]:
            minimos.append(i)
    return maximos, minimos

maximos, minimos = detectar_extremos(df["Precio"].values)
print("M√°ximos encontrados:", len(maximos))
print("M√≠nimos encontrados:", len(minimos))
```

```python

plt.figure(figsize=(12,6))
plt.plot(df["Fecha"], df["Precio"], label='Precio', color='blue')
plt.scatter(df["Fecha"].iloc[maximos], df["Precio"].iloc[maximos], color='red', label='M√°ximos')
plt.scatter(df["Fecha"].iloc[minimos], df["Precio"].iloc[minimos], color='green', label='M√≠nimos')
plt.title("Detecci√≥n de M√°ximos y M√≠nimos Locales")
plt.xlabel("Fecha")
plt.ylabel("Precio")
plt.legend()
plt.grid(True)
plt.show()
```

A trav√©s del an√°lisis de la curva de precios:

- Se identificaron **picos locales** como posibles zonas de **venta**.
- Se marcaron **valles locales** como posibles zonas de **compra**.
